```python
import pandas as pd
import numpy as np

def getMeansAndSD(filename, worksheet):
    """
    Read an Excel file and calculate means and standard
    deviations for blocks of 3 values.
    """
    means = []
    stds = []
    df = pd.read_excel(filename, worksheet)
    
    row = -1
    
    while True:
        row += 1
        
        if row % 3 == 2:
            continue
        
        if np.isnan(df.iloc[row, 0]):
            break
        
        col = 0
        while not np.isnan(df.iloc[row, col]):
            values = df.iloc[row, col:col + 3]
            means.append(values.mean())
            stds.append(values.std())
            col += 3
            
    return pd.DataFrame({
            'means': means,
            'stds': stds,
        })
```
